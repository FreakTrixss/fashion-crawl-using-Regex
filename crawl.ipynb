```python
content = tst_soup.find_all('p')
''' for par in content :
    if par.text == '' :
        pass
    else :
        txt = par.text
        break

pprint(type(txt))
pprint(txt) '''
data = {}
body = ""
for p2 in content:
    body += p2.text.strip()
    # data = {}


pattern_name = r'((?:\w+ ?){1,3}) is'
match = re.search(pattern_name, body)

if match:
    name = match.group(1)
    data['name'] = name


pattern_age = r"(\d+) years old"
match = re.search(pattern_age, body)

if match:
    data['age'] = match.group(1) + 'years old'


pattern_gender = r'\b(he|she|his|her)\b'

matches = re.findall(pattern_gender, body, re.IGNORECASE)
    # print(matches)
if matches:
    gender = "Female" if any(
        match.lower() in ['she', 'her'] for match in matches) else "Male"
    data['gender'] = gender


pattern_country = r"from (?:the)?\s*([a-zA-Z\s]+[^.])"
match = re.search(pattern_country, body)

if match:
    data['country/city'] = match.group(1)

pattern_national = r"(\w+) by nationality"
nationality = re.search(pattern_national, body, re.IGNORECASE)

if nationality:
    data['nationality'] = nationality.group(1)


birthday = re.findall(
        r'(?:\d{1,2} )?(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* (?:\d{1,2}, )?(?:\d{4})?', body)

if birthday:
    data['birthday'] = birthday[0]
else :
    data['birthday'] = 'None'
 

pattern_sign = r"Zodiac sign is (\w+)"
sign = re.search(pattern_sign, body)

if sign:
    data['zodiac sign'] = sign.group(1)


pattern_height = r"is approximately (\d?\d+??) inches"
height = re.search(pattern_height, body)

if height:
    data['height'] = height.group(1) + 'inches'


pattern_weight = r"weighs around (\d+ \w+)"
weight = re.search(pattern_weight, body)

if weight:
    data['weight'] = weight.group(1)


pattern = r"has (\w+)? eyes and (\w+) hair"
eyes_color = re.search(pattern, body)
hair_color = re.search(pattern, body)
if eyes_color:
    data['eyes-color'] = eyes_color.group(1)

if hair_color:
    data['hair-color'] = hair_color.group(2)

# pprint(body)

pprint(data)
```

```python
info = []

for i in range(1025):
    url = f"https://wikifamouspeople.com/category/instagram-star/page/{i+1}/"
    request = requests.get(url)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')

    articles = soup.find_all(
        'article', attrs={'class': 'latestPost excerpt first'})
    urls = []

    for item in articles:
        link = item.find('a')
        # print(link['href'])
        urls.append(link['href'])

    articles2 = soup.find_all('article', attrs={'class': 'latestPost excerpt'})

    for item in articles2:
        link = item.find('a')
        # print(link['href'])
        urls.append(link['href'])

    articles3 = soup.find_all(
        'article', attrs={'class': 'latestPost excerpt last'})

    for item in articles3:
        link = item.find('a')
        # print(link['href'])
        urls.append(link['href'])

    for link in urls:
        # print(link)
        data = {}
        data['link'] = link
        new_request = requests.get(link)
        tst_htm = new_request.text
        tst_soup = BeautifulSoup(tst_htm, 'html.parser')

        pars = tst_soup.find_all('p')

        # full_body = ""

        # for p in img :
        # ps = p.find_all('p')
        # for t in ps :
        # print(t.text)
        # full_body += p.text
        # print(full_body)

        head = tst_soup.find(
            'h1', attrs={'class': 'title single-title entry-title'})
        # print(head.text)

        if head:
            data['title'] = head.text.strip()
        else:
            data['title'] = 'None'
        date = tst_soup.find(
            'span', attrs={'class': 'thetime date updated'})
        # print(date.text)
        if date:
            date2 = date.find('span')
            data['date'] = date.text.strip()
        else:
            data['date'] = 'None'

        author = tst_soup.find(
            'span', attrs={'class': 'theauthor'})
        # print(author.text)

        if author:
            author2 = author.find('span')
            data['author'] = author2.text.strip()
        else:
            data['author'] = 'None'

            # data['body'] = full_body
        body = ""
        for p in pars:
            body += p.text.strip()

        if '\xa0' in body:
            body = body.replace('\xa0', ' ')
        # pattern_name = r'((?:\w+[_]? ?){1,3}) is'
        pattern_name = r'\b((?:(?! is\b)\w+)(?: (?!(?:is|from)\b)\w+){0,2})\b'
        pattern_name = r'\b((?:(?! is\b)\w+)\.?\w+.?\s?(?!is|from)\w+?(?: (?!(?:is|from)\b)\w+){0,2})\b'
        match = re.search(pattern_name, body)

        if match:
            name = match.group(1)
            data['name'] = name

        pattern_age = r"(\d+) years old"
        match = re.search(pattern_age, body)

        if match:
            data['age'] = match.group(1) + ' years old'

        pattern_gender = r'\b(he|she|his|her)\b'

        matches = re.findall(pattern_gender, body, re.IGNORECASE)
        # print(matches)
        if matches:
            gender = "Female" if any(
                match.lower() in ['she', 'her'] for match in matches) else "Male"
            data['gender'] = gender
        else:
            data['gender'] = 'None'

        pattern_country = r"from (?:the)?\s*([a-zA-Z\s*?]+[^.])"
        match = re.search(pattern_country, body, re.I)

        if match:
            data['country/city'] = match.group(1)
        else:
            data['country/city'] = 'None'

        pattern_national = r"(\w+) by nationality"
        nationality = re.search(pattern_national, body, re.IGNORECASE)

        if nationality:
            data['nationality'] = nationality.group(1)
        else:
            data['nationality'] = 'None'
        # birthday = re.findall(
        #         r'(?:\d{1,2} )?(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* (?:\d{1,2}, )?\d{4}', body , re.I)

        birthday = re.search(
            r'(\d{1,2}(?:st|nd|rd|th)? [A-Za-z]+,? \d{4}|[A-Za-z]+ \d{1,2}, \d{4})', body, re.I)

        if birthday:
            data['birthday'] = birthday.group(1)
        else:
            data['birthday'] = 'None'

        pattern_sign = r"zodiac sign is (\w+)"
        sign = re.search(pattern_sign, body, re.I)

        if sign:
            data['zodiac sign'] = sign.group(1)
        else:
            data['zodiac sign'] = 'None'

        pattern_height = r'(\d[\'\"]\s?\d{1,2}\s?["]?) inches?'
        height = re.search(pattern_height, body, re.I)

        if height:
            data['height'] = height.group(1) + ' inches'
        else:
            data['height'] = 'None'

        # pattern_weight = r"weighs around (\d+ \w+)"
        pattern_weight = r"weighs? around (\d+\s?\w+)"
        weight = re.search(pattern_weight, body)

        if weight:
            data['weight'] = weight.group(1)
        else:
            data['weight'] = 'None'

        pattern = r"has (\w+)? eyes and (\w+) hair"
        eyes_color = re.search(pattern, body)
        hair_color = re.search(pattern, body)
        if eyes_color:
            data['eyes-color'] = eyes_color.group(1)
        elif re.search(r'\b(\w+)\s?eyes\b', body, re.I):
            data['eyes-color'] = re.search(r'\b(\w+)\s?eyes\b',
                                           body, re.I).group(1)
        else:
            data['eyes-color'] = 'None'

        if hair_color:
            data['hair-color'] = hair_color.group(2)
        elif re.search(r"(\w+) hair\b", body, re.I):
            data['hair-color'] = re.search(r"(\w+) hair\b",
                                           body, re.I).group(1)
        else:
            data['hair-color'] = 'None'

        info.append(data)
```
